```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_classification

# تولید داده‌های نمونه
X, y = make_classification(n_samples=225, n_features=20, n_informative=15, n_classes=15, random_state=42)

# اجرای T-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# رسم نمودار
plt.figure(figsize=(10, 8))
for i in range(15):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=f'Class {i+1}')
plt.legend()
plt.title('T-SNE for 15 Classes')
plt.xlabel('T-SNE Component 1')
plt.ylabel('T-SNE Component 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# تولید داده‌های نمونه
X, y = make_classification(n_samples=225, n_features=20, n_informative=15, n_classes=15, random_state=42)

# تقسیم داده‌ها به مجموعه‌های آموزشی و آزمایشی
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ساخت مدل طبقه‌بندی
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# پیش‌بینی و محاسبه دقت
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')

# اجرای T-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# رسم نمودار
plt.figure(figsize=(10, 8))
for i in range(15):
    plt.scatter(X_tsne[y == i, 0], X_tsne[y == i, 1], label=f'Class {i+1}')
plt.legend()
plt.title('T-SNE for 15 Classes')
plt.xlabel('T-SNE Component 1')
plt.ylabel('T-SNE Component 2')
plt.show()
```
