```python
#pip install vit-pytorch
import torch
import torch.nn as nn
import torch.nn.functional as F
from vit_pytorch import ViT
from transformers import ViTModel, ViTConfig
from torch.optim import Adam
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

# بارگذاری مدل ViT
config = ViTConfig.from_pretrained('google/vit-base-patch16-224')
model = ViTModel(config)
device = torch.cuda.is_available()
if device:
    model.cuda()
input_embedding = test_embeddings 
labels = test_labels

print(input_embedding.shape) 
# Check the type and shape of input_embedding
print(type(input_embedding))
print(input_embedding.shape)

# Convert to tensor if necessary
if not isinstance(input_embedding, torch.Tensor):
    input_embedding = torch.tensor(input_embedding)
    input_embedding = input_embedding.to('cuda')
    print(type(input_embedding))
    print(input_embedding.shape)

# اضافه کردن پدینگ به ورودی
padding_size = 64  # مقدار پدینگ مورد نیاز برای هر طرف
input_embedding = F.pad(input_embedding, (0, padding_size))

print(input_embedding.shape)  # بررسی شکل جدید ورودی

# تغییر شکل به [batch_size, channels, height, width]
input_embedding = input_embedding.view(4122, 1, 1, 1344)

# تغییر اندازه به 224x224
input_embedding = F.interpolate(input_embedding, size=(224, 224), mode='bilinear', align_corners=False)

# تغییر شکل به [batch_size, channels, height, width]
input_embedding = input_embedding.view(4122, 1, 224, 224)

# تکرار کانال‌ها برای تبدیل به 3 کانال
input_embedding = input_embedding.repeat(1, 3, 1, 1)

print(input_embedding.shape)  # بررسی شکل جدید ورودی

# تعریف لایه طبقه‌بندی
class ViTClassifier(nn.Module):
    def __init__(self, vit_model):
        super(ViTClassifier, self).__init__()
        self.vit = vit_model
        self.fc = nn.Linear(config.hidden_size, 2)  # فرض کنید دو کلاس داریم

    def forward(self, x):
        batch_size = x.size(0)
        channels = 3
        height = 224
        width = 224

        # Ensure the dimensions are valid
        if batch_size * channels * height * width == x.numel():
            x = x.view(batch_size, channels, height, width)
        else:
            raise ValueError("Invalid dimensions for reshaping")

        outputs = self.vit(x)
        return self.fc(outputs.last_hidden_state[:, 0, :])  # استفاده از توکن CLS

# مدل را ایجاد و تنظیمات
classifier = ViTClassifier(model)
optimizer = Adam(classifier.parameters(), lr=1e-5)
criterion = nn.CrossEntropyLoss()

# آموزش مدل
classifier.train()
for epoch in range(5):  # تعداد دوره‌ها را تنظیم کنید
    optimizer.zero_grad()
    outputs = classifier(input_embedding)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
    print(f'Epoch {epoch + 1}, Loss: {loss.item()}')

# ارزیابی مدل
classifier.eval()
with torch.no_grad():
    predictions = classifier(input_embedding)
    predicted_labels = torch.argmax(predictions, dim=1)

# محاسبه دقت و F1 score
accuracy = accuracy_score(labels.numpy(), predicted_labels.numpy())
f1 = f1_score(labels.numpy(), predicted_labels.numpy(), average='weighted')

print(f'Accuracy: {accuracy:.4f}')
print(f'F1 Score: {f1:.4f}')
 
 

# تبدیل embedding‌ها به تصاویر مصنوعی
#def embeddings_to_images(embeddings):
#    images = []
#    for embedding in embeddings:
#        image = np.reshape(embedding, (1, 1, 1280))  # تبدیل به یک تصویر مصنوعی
#        images.append(image)
#    return np.array(images)


 
# ارزیابی نتایج
#from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

#accuracy = accuracy_score(test_labels, predictions)
#precision = precision_score(test_labels, predictions, average='macro')
#recall = recall_score(test_labels, predictions, average='macro')
#f1 = f1_score(test_labels, predictions, average='macro')

#print(f'Accuracy: {accuracy}')
#print(f'Precision: {precision}')
#print(f'Recall: {recall}')
#print(f'F1 Score: {f1}')
#    
```
